```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best",shadow="True")
plt.grid()
```

```python
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos= np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.title("Companies revenue and profit plot")
plt.xlabel("Company")
plt.ylabel("Revenue(bln)")
plt.bar(xpos,revenue,label="Revenue")
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos,company)
plt.title("Companies revenue and profit plot")
plt.xlabel("Company")
plt.ylabel("Revenue(bln)")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar,rwidth=0.7)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150])
```

```python

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],rwidth=0.95,label=["men","women"])
plt.legend()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%",radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%",radius=2,explode=[0,0,0.2,0.3,0])
plt.savefig("pie.png" ,bbox_inches="tight", pad_inches=5)
plt.savefig("pie.pdf" ,bbox_inches="tight", pad_inches=5)
plt.show()
```
